### Задача 1
Напишем программу для проверки логина и пароля

1) логин и пароль пользователя хранится в переменной

2) будем давать пользователю k попыток на ввод логина и пароля

3) если после трех попыток вверного ввода не было, то заблокируем вход - вывести на экран сообщение "Вход заблокирован"

4) если пароль и логин были введены верно, то вывести на экран сообщение "Добро пожаловать!"

```python
correct_login = 'admin'  # snake_case
correct_password = "admin"

# CamelCase --> классы и константы
```

```python
'select * from table where date=\'2022-09-14\''  # backtick `
```

```python
# флаг равен False, когда вход заблокирован
flag = False
N_TRIES = 3

for step in range(N_TRIES):
    print(step)
    
    login = input('Введите логин')
    password = input('Введите password')
    
    if login == correct_login and password == correct_password:
        flag = True
        break
        
        
if flag:  # переменная не равна None, '', {}, [], 0, False
    print('Добро пожаловать!')
else:
    print('Вход заблокирован')
        
```

### Задача 4
Посчитаем средний возраст в коллективе

```python
people = {1: {'name': 'Oleg', 'age': '29', 'sex': 'Male'},
          2: {'name': 'Kate', 'age': '21', 'sex': 'Female'},
          3: {'name': 'Liza', 'age': '24', 'sex': 'Female'},
          4: {'name': 'Pavel', 'age': '36', 'sex': 'Male'}}
```

```python
age_sum = 0
age_count = 0  # это совпадет с len(people)

for value in people.values():
    # print(value, value['age'], type(value['age']))
    # print(value, age, type(age))
    
    age = int(value['age'])
    print(age)
    
    age_sum += age
    age_count += 1
    
    # break
    
print(age_sum / age_count)
```

### Из финального варианта решения удаляйте все закомментированные строки

Дан список с визитами по городам и странам. 
Напишем код, который возвращает отфильтрованный список `geo_logs`, содержащий только визиты из России.


```python
geo_logs = [
    {'visit1': ['Москва', 'Россия']},
    {'visit2': ['Дели', 'Индия']},
    {'visit3': ['Владимир', 'Россия']},
    {'visit4': ['Лиссабон', 'Португалия']},
    {'visit5': ['Париж', 'Франция']},
    {'visit7': ['Тула', 'Россия']},
    {'visit9': ['Курск', 'Россия']},
    {'visit10': ['Архангельск', 'Россия']},
    None
]
```

```python
list(geo_logs[0].keys())[0]
```

```python
list(geo_logs[0].values())
```

```python
# вычисление list(geo_logs[0].values()) дает
# [['Москва', 'Россия']]
```

```python
filtered_geo_logs = []

for record in geo_logs:
    # dict_key = list(record.keys())[0]
    # country = record[dict_key][1]
    
    if record:
        # вычисление list(geo_logs[0].values()) дает
        # [['Москва', 'Россия']]
        country = list(record.values())[0][1]

        if country == 'Россия':
            filtered_geo_logs.append(record)
    
filtered_geo_logs
```

```python
# откуда взять токен https://dev.vk.com/api/getting-started#Авторизация%20пользователя
token = open('token.txt').read()
```

Поищем сообщества в социальной сети

```python
params = {
    'q': 'python',
    'access_token': token,
    'v':'5.131',
    'count': 300
}
req = requests.get('https://api.vk.com/method/groups.search', params).json()
```

```python
print(req)
```

```python
pprint(req)
```

```python
target_groups = req['response']['items']
target_groups
```

```python
# преобразуем список всех id в строку (в таком виде принимает данные параметр fields)
target_group_ids = ','.join([str(group['id']) for group in target_groups])
target_group_ids
```

```python
params = {
    'access_token': token,
    'v':'5.131',
    'group_ids': target_group_ids,
    'fields':  'members_count,activity,description'

}
req = requests.get('https://api.vk.com/method/groups.getById', params)

pprint(req.json())
```
